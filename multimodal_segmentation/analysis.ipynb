```python
import matplotlib.pyplot as plt
```

```python
train_acc_multimodal = [74.28, 94.09, 96.06, 97.78, 98.38]
val_acc_multimodal = [91.75, 96.93, 97.88, 98.58, 97.41]

train_acc_simple_mri = [76.06, 89.63, 93.74, 95.80, 97.14]
val_acc_simple_mri = [86.53, 87.73, 93.20, 96.00, 96.53]
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(train_acc_multimodal, label='Multimodal')
ax.plot(train_acc_simple_mri, label='Simple MRI')
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.set_title('Training Accuracy')
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(val_acc_multimodal, label='Multimodal')
ax.plot(val_acc_simple_mri, label='Simple MRI')
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.set_title('Training Accuracy')
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(val_acc_multimodal, label='Multimodal')
ax.plot(val_acc_simple_mri, label='Simple MRI')
ax.set_xlim(1, 5)
ax.set_ylim(0, 100)
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.set_title('Training Accuracy')
ax.legend()
```
